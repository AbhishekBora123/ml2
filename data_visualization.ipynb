```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('fmri')
```

```python
data.head(10)
```

# **LINE PLOT**

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='timepoint',y='signal',data=data,hue='event',style='event',markers=True)
plt.show()
```

## **SCATTER PLOT**

```python
from sklearn.datasets import load_iris
```

```python
data3=load_iris()
```

```python
data3
```

```python
import numpy as np
```

```python
data3 = pd.DataFrame(data = np.c_[data3['data'],data3['target']],columns=data3['feature_names']+['target'])
```

```python
data3.head(10)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='sepal length (cm)',y='petal length (cm)',data=data3,hue='target')
plt.show()
```

## **HISTOGRAM DISTPLOT**

```python
data4=pd.read_csv('/content/diamonds.csv')
```

```python
data4.head(10)
```

```python
sns.distplot(data4['price'])
plt.show()
```

```python
sns.distplot(data4['price'],hist=False,color='g')
plt.show()
```

```python
sns.distplot(data4['price'],kde=False,color='b')
plt.show()
```

```python
sns.distplot(data4['price'],kde=False,color='g',bins=6)
plt.show()
```

```python
sns.distplot(data4['price'],kde=False,color='g',bins=6,vertical=True)
plt.show()
```
